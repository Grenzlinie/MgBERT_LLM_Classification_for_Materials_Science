```python
import pandas as pd
data = pd.read_excel('../classification_models/bert_acc_result.xlsx')
import matplotlib.pyplot as plt
```

```python
epoch = data['epoch'].tolist()
test_set_accuracy_series1 = [x * 100 for x in data['matscibert_accuracy_b26'].tolist()]
test_set_accuracy_series2 = [x * 100 for x in data['bert_accuracy_b26'].tolist()]
test_set_accuracy_series3 = [x * 100 for x in data['longformer_accuracy'].tolist()]

plt.figure(figsize=(9, 6))

plt.plot(epoch, test_set_accuracy_series1, color='skyblue', label='MatsciBert', linewidth=3.0)
plt.plot(epoch, test_set_accuracy_series2, color='lightcoral', label='BERT', linewidth=3.0)
plt.plot(epoch, test_set_accuracy_series3, color='lightgreen', label='Longformer', linewidth=3.0)
plt.xlabel('Epoch', fontsize=18)
plt.ylabel('Testset_accuracy (%)', fontsize=18)
plt.title('Model accuracy variation with epoch', fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
# 调整x轴和y轴的刻度线粗细
plt.tick_params(axis='x', width=2)
plt.tick_params(axis='y', width=2)
# 调整整个折线图画框的粗细
plt.gca().spines['top'].set_linewidth(2)       # 上边框
plt.gca().spines['bottom'].set_linewidth(2)    # 下边框
plt.gca().spines['left'].set_linewidth(2)      # 左边框
plt.gca().spines['right'].set_linewidth(2)     # 右边框

plt.legend(fontsize=18)
plt.annotate(f'{test_set_accuracy_series1[-1]:.1f}', (epoch[-1], test_set_accuracy_series1[-1]), textcoords="offset points", xytext=(-4,-5), ha='center', fontsize=18)
plt.annotate(f'{test_set_accuracy_series2[-1]:.1f}', (epoch[-1], test_set_accuracy_series2[-1]), textcoords="offset points", xytext=(-4,5), ha='center', fontsize=18)
plt.annotate(f'{test_set_accuracy_series3[-1]:.1f}', (epoch[-1], test_set_accuracy_series3[-1]), textcoords="offset points", xytext=(-4,-10), ha='center', fontsize=18)

plt.savefig('epoch-accuracy.svg', dpi=600)
plt.show()


```

```python
import numpy as np

labels = ['BERT', 'MatSciBERT', 'Longformer', 'MgBERT']
trainset_loss = [0.010, 0.005, 0.023, 0.000]
testset_loss = [0.018, 0.018, 0.034, 0.025]


fig, ax = plt.subplots()
fig.set_size_inches(10, 7)
bar_width = 0.42
index = np.arange(len(labels))
fontsize = 20
rects1 = ax.bar(index, trainset_loss, bar_width, label='trainset_loss', color='#c2e9fb', edgecolor='black', linewidth=2.5)  
rects2 = ax.bar(index + bar_width, testset_loss, bar_width, label='testset_loss', color='#a1c4fd', edgecolor='black', linewidth=2.5) 

for i, rect in enumerate(rects1):
    height = rect.get_height()
    if height == 0:
        ax.text(rect.get_x() + rect.get_width() / 2, height, f'{trainset_loss[i]:.3f}',
            ha='center', va='bottom', fontsize=fontsize)
    else:
        ax.text(rect.get_x() + rect.get_width() / 2, height-0.0008, f'{trainset_loss[i]:.3f}',
            ha='center', va='top', fontsize=fontsize)

for i, rect in enumerate(rects2):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height-0.0008, f'{testset_loss[i]:.3f}',
            ha='center', va='top', fontsize=fontsize)

ax.set_xlabel('Models', fontsize=fontsize)
ax.set_ylabel('Loss', fontsize=fontsize)
ax.set_title('Trainset and Testset Loss for Different Models', fontsize=fontsize)
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(labels)
ax.legend(fontsize=fontsize)

plt.xticks(fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
# 调整x轴和y轴的刻度线粗细
plt.tick_params(axis='x', width=2)
plt.tick_params(axis='y', width=2)
# 调整整个折线图画框的粗细
plt.gca().spines['top'].set_linewidth(2.5)       # 上边框
plt.gca().spines['bottom'].set_linewidth(2.5)    # 下边框
plt.gca().spines['left'].set_linewidth(2.5)      # 左边框
plt.gca().spines['right'].set_linewidth(2.5)     # 右边框


fig.tight_layout()
plt.savefig('final_compare_loss.svg', dpi=600)
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['BERT', 'MatSciBERT', 'Longformer', 'MgBERT']
trainset_acc = [88.8, 95.0, 88.0, 99.9]
testset_acc = [81.6, 85.0, 81.4, 88.5]

fig, ax = plt.subplots()
fig.set_size_inches(10, 7)
bar_width = 0.4
index = np.arange(len(labels))
fontsize = 20
rects1 = ax.bar(index, trainset_acc, bar_width, label='trainset_acc', color='#ffecd2', edgecolor='black', linewidth=2.5)
rects2 = ax.bar(index + bar_width, testset_acc, bar_width, label='testset_acc', color='#fcb69f', edgecolor='black', linewidth=2.5)

for i, rect in enumerate(rects1):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height-0.8, trainset_acc[i],
            ha='center', va='top', fontsize=fontsize)

for i, rect in enumerate(rects2):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height-0.8, testset_acc[i],
            ha='center', va='top', fontsize=fontsize)

ax.set_xlabel('Models', fontsize=fontsize)
ax.set_ylabel('Accuracy (%)', fontsize=fontsize)
ax.set_title('Trainset and Testset Accuracy for Different Models', fontsize=fontsize)
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(labels)
ax.legend(fontsize=fontsize, loc='upper left', ncol=2)

plt.xticks(fontsize=fontsize)
plt.xticks(fontsize=fontsize)
plt.yticks(fontsize=fontsize)
# 调整x轴和y轴的刻度线粗细
plt.tick_params(axis='x', width=2)
plt.tick_params(axis='y', width=2)
# 调整整个折线图画框的粗细
plt.gca().spines['top'].set_linewidth(2.5)       # 上边框
plt.gca().spines['bottom'].set_linewidth(2.5)    # 下边框
plt.gca().spines['left'].set_linewidth(2.5)      # 左边框
plt.gca().spines['right'].set_linewidth(2.5)     # 右边框

plt.ylim(60, 100)  # 设置y轴刻度最高是100
fig.tight_layout()
plt.savefig('final_compare_acc.svg', dpi=600)

plt.show()
```
